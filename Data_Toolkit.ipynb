1. What is NumPy, and why is it widely used in Python?
>> NumPy is a Python library for numerical computing that provides support for large, multi-dimensional arrays and mathematical operations. It is widely used because it is fast, efficient, and enables vectorized operations.

2. How does broadcasting work in NumPy?
>> Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding smaller arrays to match the larger ones without copying data.

3. What is a Pandas DataFrame?
>> A Pandas DataFrame is a two-dimensional, labeled data structure in the Pandas library. It is similar to a table in Excel, with rows and columns, and is used to store, manipulate, and analyze structured data efficiently.

4. Explain the use of the groupby() method in Pandas?
>> The groupby() method in Pandas is used to split data into groups based on one or more columns, apply operations (like sum, mean, count), and combine the results. It helps in summarizing and analyzing data efficiently.

5. Why is Seaborn preferred for statistical visualization?
>> Seaborn is preferred for statistical visualization because it provides built-in themes, attractive graphics, and easy-to-use functions for complex statistical plots like distributions, correlations, and regression analysis.

6. What are the differences between NumPy arrays and Python lists?
>> NumPy arrays and Python lists differ in performance, data type, and functionality. NumPy arrays are faster, memory-efficient, and support vectorized mathematical operations, while Python lists are slower and can store mixed data types. NumPy is mainly used for numerical and scientific computing, whereas lists are used for general-purpose programming.

7. What is a heatmap, and when should it be used?
>> A heatmap is a data visualization technique that uses colors to represent values in a matrix or dataset. It helps in quickly identifying patterns, trends, and correlations. Heatmaps are commonly used in data analysis, especially for correlation matrices, website user behavior, and performance comparisons.

8. What does the term “vectorized operation” mean in NumPy?
>> In NumPy, a vectorized operation means performing operations on entire arrays at once without using explicit loops. It makes computations faster and more efficient.

9. How does Matplotlib differ from Plotly?
>> Matplotlib is a static plotting library used for basic and customizable visualizations, while Plotly is used to create interactive and dynamic charts with zooming, hovering, and web integration.

10. What is the significance of hierarchical indexing in Pandas?
>> Hierarchical indexing also called MultiIndex in Pandas allows multiple levels of indexing on rows or columns. It helps represent higher-dimensional data in a lower-dimensional (tabular) format and makes complex data analysis easier.

11. What is the role of Seaborn’s pairplot() function?
>> The pairplot() function in Seaborn is used to visualize relationships between multiple variables in a dataset by plotting pairwise scatter plots and distributions in a single grid.

12. What is the purpose of the describe() function in Pandas?
>> The describe() function in Pandas is used to generate summary statistics of a dataset, such as count, mean, standard deviation, minimum, maximum, and quartiles.

13. Why is handling missing data important in Pandas?
>> Handling missing data is important in Pandas because incomplete data can lead to incorrect analysis and misleading results. Missing values may distort statistical calculations like mean and standard deviation, and can even cause errors in data processing or machine learning models. By identifying and treating missing data using methods like dropna() or fillna(), we ensure the dataset remains clean and reliable. Proper handling improves data quality, supports accurate insights, and enables better decision-making in data analysis.

It allows users to create dynamic charts with features like zooming, panning, and hover tooltips, making data exploration easier. Plotly supports web-based outputs, which are ideal for dashboards and presentations. It integrates well with Python and Pandas, enabling quick visualization of datasets. 15. How does NumPy handle multidimensional arrays?
>> NumPy handles multidimensional arrays using a powerful data structure called the ndarray. It allows creation and manipulation of arrays with multiple dimensions (e.g., 2D, 3D) in a structured and efficient way. Each array has attributes like shape, size, and dimensions that help manage data easily. NumPy supports operations across axes, slicing, and indexing for quick access to elements. It also enables vectorized computations on entire arrays, making multidimensional data processing fast, memory-efficient, and suitable for scientific and analytical tasks.

17. Explain the difference between apply() and map() in Pandas?
>> In Pandas, map() is used for element-wise operations on a single column (Series), such as replacing or transforming values using a function, dictionary, or mapping. It works only on Series. In contrast, apply() is more flexible and can be used on both Series and DataFrames to apply functions across rows or columns. While map() is simpler and faster for basic transformations, apply() is preferred for more complex operations and custom logic on larger datasets.

18. What are some advanced features of NumPy?
>> NumPy provides several advanced features for efficient numerical computing. It supports broadcasting, allowing operations on arrays of different shapes. It offers powerful vectorized operations for fast computations without loops. NumPy includes advanced indexing and slicing techniques for flexible data access. It also provides functions for linear algebra, random number generation, and Fourier transforms. Additionally, NumPy arrays are memory-efficient and integrate well with libraries like Pandas and SciPy, making it a core tool for scientific computing and data analysis.

19. How does Pandas simplify time series analysis?
>> Pandas simplifies time series analysis by providing powerful tools to work with date and time data. It offers built-in datetime indexing, allowing easy filtering, slicing, and alignment of time-based data. Functions like resampling help in changing data frequency (daily, monthly), while rolling and expanding windows support trend and moving average analysis. Pandas also handles missing time data efficiently and enables time zone conversions. These features make it easier to analyze trends, patterns, and seasonal variations in time series datasets.

20. What is the role of a pivot table in Pandas?
>> In Pandas, a pivot table is used to summarize and reorganize data for better analysis. It helps transform large datasets into a structured format by grouping values based on categories and applying aggregation functions like sum, mean, or count. Pivot tables make it easy to compare data across different dimensions (e.g., sales by region and product). They improve readability and support quick insights, similar to Excel pivot tables, making them highly useful for data exploration and reporting.

21.  Why is NumPy’s array slicing faster than Python’s list slicing?
>> NumPy array slicing is faster than Python list slicing because it uses views instead of copies. When slicing a NumPy array, it typically creates a new view that references the original data without duplicating it, making the operation memory-efficient and fast. In contrast, Python list slicing creates a new copy of the data, which takes more time and memory. Additionally, NumPy is implemented in optimized C code, further improving performance for large datasets and numerical computations.

22. What are some common use cases for Seaborn?
>> Seaborn is widely used for statistical data visualization and exploratory data analysis (EDA). Common use cases include visualizing data distributions using histograms and KDE plots, analyzing relationships between variables with scatter plots and pairplots, and understanding correlations through heatmaps. It is also used for comparing categories using boxplots and violin plots. Seaborn helps identify trends, patterns, and outliers in datasets, making it an essential tool for data analysts and data scientists.





```python
#1. How do you create a 2D NumPy array and calculate the sum of each row?
#In NumPy, a 2D array can be created using the array() function with nested lists.
#To calculate the sum of each row, the sum() function is used with axis=1, which performs row-wise addition.

import numpy as np

arr = np.array([[11, 22, 33],
                [44, 55, 66]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd

data = {'Marks': [80, 85, 91, 56]}
df = pd.DataFrame(data)

mean_value = df['Marks'].mean()
print(mean_value)
```

```python
#3. Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot")
plt.show()
```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'A':[1,2,3], 'B':[4,5,6], 'C':[7,8,9]}
df = pd.DataFrame(data)

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
#8. Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]
y3 = [25, 35, 45, 55]


plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.plot(x, y3, label="Line 3")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()
plt.show()
```
